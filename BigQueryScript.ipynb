```python
# This is what was used to query in BigQuery
# This is how the HackerNews was pulled

SELECT id, timestamp, text # Pull the id, timestamp and text
FROM `bigquery-public-data.hacker_news.full` # Pull data from Hacker News
WHERE timestamp >= '2022-01-01' # Start date
  AND EXTRACT(DAYOFWEEK FROM timestamp) NOT IN (1, 7) # Do not include weekends
  AND DATE(timestamp) NOT IN ( # Do not include these holidays
    DATE '2023-01-16', DATE '2023-02-20', DATE '2023-04-07', DATE '2023-05-29',
    DATE '2023-06-19', DATE '2023-07-04', DATE '2023-09-04', DATE '2023-11-23',
    DATE '2023-12-25', DATE '2024-01-01', DATE '2024-01-15', DATE '2024-02-19',
    DATE '2024-03-29', DATE '2024-05-27', DATE '2024-06-19', DATE '2024-07-04',
    DATE '2024-09-02', DATE '2024-11-28', DATE '2024-12-25'
  )
  AND ( # Select articles where these companies were mentioned
    LOWER(text) LIKE '%aapl%' OR LOWER(text) LIKE '%apple%' OR
    LOWER(text) LIKE '%msft%' OR LOWER(text) LIKE '%microsoft%' OR
    LOWER(text) LIKE '%nvda%' OR LOWER(text) LIKE '%nvidia%' OR
    LOWER(text) LIKE '%amzn%' OR LOWER(text) LIKE '%amazon%' OR
    LOWER(text) LIKE '%meta%' OR LOWER(text) LIKE '%facebook%' OR LOWER(text) LIKE '%meta platforms%' OR
    LOWER(text) LIKE '%brk.b%' OR LOWER(text) LIKE '%berkshire%' OR LOWER(text) LIKE '%buffett%' OR LOWER(text) LIKE '%berkshire hathaway%' OR
    LOWER(text) LIKE '%googl%' OR LOWER(text) LIKE '%alphabet%' OR LOWER(text) LIKE '%google%' OR
    LOWER(text) LIKE '%goog%' OR LOWER(text) LIKE '%alphabet%' OR LOWER(text) LIKE '%google%' OR
    LOWER(text) LIKE '%tsla%' OR LOWER(text) LIKE '%tesla%' OR LOWER(text) LIKE '%elon%' OR
    LOWER(text) LIKE '%jpm%' OR LOWER(text) LIKE '%jpmorgan%' OR LOWER(text) LIKE '%jp morgan%' OR
    LOWER(text) LIKE '%unh%' OR LOWER(text) LIKE '%unitedhealth%' OR
    LOWER(text) LIKE '%archer%' OR LOWER(text) LIKE '%archer aviation%' OR LOWER(text) LIKE '%achr%' OR
    LOWER(text) LIKE '%cost%' OR LOWER(text) LIKE '%costco%' OR
    LOWER(text) LIKE '%nflx%' OR LOWER(text) LIKE '%netflix%' OR
    LOWER(text) LIKE '%wmt%' OR LOWER(text) LIKE '%walmart%' OR
    LOWER(text) LIKE '%pg%' OR LOWER(text) LIKE '%procter%' OR LOWER(text) LIKE '%procter and gamble%' OR
    LOWER(text) LIKE '%jnj%' OR LOWER(text) LIKE '%johnson%' OR LOWER(text) LIKE '%johnson & johnson%' OR
    LOWER(text) LIKE '%hd%' OR LOWER(text) LIKE '%home depot%' OR
    LOWER(text) LIKE '%abbv%' OR LOWER(text) LIKE '%abbvie%' OR
    LOWER(text) LIKE '%ko%' OR LOWER(text) LIKE '%coca cola%' OR LOWER(text) LIKE '%coke%' OR
    LOWER(text) LIKE '%crm%' OR LOWER(text) LIKE '%salesforce%' OR
    LOWER(text) LIKE '%pm%' OR LOWER(text) LIKE '%philip morris%' OR
    LOWER(text) LIKE '%cvx%' OR LOWER(text) LIKE '%chevron%' OR
    LOWER(text) LIKE '%crwd%' OR LOWER(text) LIKE '%crowdstrike%' OR
    LOWER(text) LIKE '%mcd%' OR LOWER(text) LIKE '%mcdonalds%' OR
    LOWER(text) LIKE '%orcl%' OR LOWER(text) LIKE '%oracle%' OR
    LOWER(text) LIKE '%abt%' OR LOWER(text) LIKE '%abbott%' OR
    LOWER(text) LIKE '%ibm%' OR LOWER(text) LIKE '%big blue%' OR
    LOWER(text) LIKE '%wfc%' OR LOWER(text) LIKE '%wells fargo%' OR
    LOWER(text) LIKE '%pep%' OR LOWER(text) LIKE '%pepsi%' OR LOWER(text) LIKE '%pepsico%' OR
    LOWER(text) LIKE '%mrk%' OR LOWER(text) LIKE '%merck%' OR
    LOWER(text) LIKE '%pltr%' OR LOWER(text) LIKE '%palantir%' OR
    LOWER(text) LIKE '%vz%' OR LOWER(text) LIKE '%verizon%' OR
    LOWER(text) LIKE '%acn%' OR LOWER(text) LIKE '%accenture%' OR
    LOWER(text) LIKE '%isrg%' OR LOWER(text) LIKE '%intuitive surgical%' OR
    LOWER(text) LIKE '%rtx%' OR LOWER(text) LIKE '%raytheon%' OR LOWER(text) LIKE '%rtx corp%' OR
    LOWER(text) LIKE '%tmo%' OR LOWER(text) LIKE '%thermo fisher%' OR
    LOWER(text) LIKE '%intu%' OR LOWER(text) LIKE '%intuit%' OR
    LOWER(text) LIKE '%pgr%' OR LOWER(text) LIKE '%progressive%'
  )
  AND ( # Select articles where the companies were mentioned and one of the financial terms to ensure it is relevant
    LOWER(title) LIKE '%stock%' OR LOWER(text) LIKE '%stock%' OR
    LOWER(title) LIKE '%finance%' OR LOWER(text) LIKE '%finance%' OR
    LOWER(title) LIKE '%investment%' OR LOWER(text) LIKE '%investment%' OR
    LOWER(title) LIKE '%market%' OR LOWER(text) LIKE '%market%' OR
    LOWER(title) LIKE '%price%' OR LOWER(text) LIKE '%price%' OR
    LOWER(title) LIKE '%volatility%' OR LOWER(text) LIKE '%volatility%' OR
    LOWER(title) LIKE '%dividend%' OR LOWER(text) LIKE '%dividend%' OR
    LOWER(title) LIKE '%earnings%' OR LOWER(text) LIKE '%earnings%' OR
    LOWER(title) LIKE '%trading%' OR LOWER(text) LIKE '%trading%' OR
    LOWER(title) LIKE '%shares%' OR LOWER(text) LIKE '%shares%' OR
    LOWER(title) LIKE '%equity%' OR LOWER(text) LIKE '%equity%' OR
    LOWER(title) LIKE '%portfolio%' OR LOWER(text) LIKE '%portfolio%' OR
    LOWER(title) LIKE '%returns%' OR LOWER(text) LIKE '%returns%' OR
    LOWER(title) LIKE '%profit%' OR LOWER(text) LIKE '%profit%' OR
    LOWER(title) LIKE '%growth%' OR LOWER(text) LIKE '%growth%' OR
    LOWER(title) LIKE '%revenue%' OR LOWER(text) LIKE '%revenue%' OR
    LOWER(title) LIKE '%trade%' OR LOWER(text) LIKE '%trade%'
  )
ORDER BY timestamp DESC # Order in descending order
```
